```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Read the data
df = pd.read_csv(path, sep=" ")
```

```python
df
```

```python
train_list, test_list, val_list = [], [], []

summary = []  # To store per-sequence stats

for seq, group in df.groupby("SEQUENCE"):
    seq_name = seq[:15] + "..." if len(seq) > 15 else seq
    print(f"\nProcessing SEQUENCE: {seq_name} (n={len(group)})")

    # Count label distribution
    label_counts = group["LABEL"].value_counts().to_dict()

    try:
        # Split into train (80%) and temp (20%)
        train_grp, temp_grp = train_test_split(
            group,
            test_size=0.2,
            random_state=42,
            stratify=group["LABEL"] if group["LABEL"].nunique() > 1 else None
        )

        # Split temp into test (10%) and val (10%)
        test_grp, val_grp = train_test_split(
            temp_grp,
            test_size=0.5,
            random_state=42,
            stratify=temp_grp["LABEL"] if temp_grp["LABEL"].nunique() > 1 else None
        )

        stratified = True

    except ValueError as e:
        print(f"‚ö†Ô∏è Stratified split failed: {e}. Falling back to random split.")
        train_grp, temp_grp = train_test_split(group, test_size=0.2, random_state=42, shuffle=True)
        test_grp, val_grp = train_test_split(temp_grp, test_size=0.5, random_state=42, shuffle=True)
        stratified = False

    # Save splits
    train_list.append(train_grp)
    test_list.append(test_grp)
    val_list.append(val_grp)

    # Store summary info
    summary.append({
        "Sequence (short)": seq_name,
        "Total Samples": len(group),
        "Label Dist": label_counts,
        "Train Size": len(train_grp),
        "Test Size": len(test_grp),
        "Val Size": len(val_grp),
        "Stratified?": stratified
    })

# Combine all
train_df = pd.concat(train_list).reset_index(drop=True)
test_df = pd.concat(test_list).reset_index(drop=True)
val_df = pd.concat(val_list).reset_index(drop=True)

# Save
train_df.to_csv("./dataset/b_cancer/splitted/train.csv", index=False, header=["SMILES","Target Sequence","Label"])
test_df.to_csv("./dataset/b_cancer/splitted/test.csv", index=False, header=["SMILES","Target Sequence","Label"])
val_df.to_csv("./dataset/b_cancer/splitted/val.csv", index=False, header=["SMILES","Target Sequence","Label"])

# Summary DataFrame
summary_df = pd.DataFrame(summary)
print("\n‚úÖ Split completed successfully!")
print(f"Train: {len(train_df)} rows")
print(f"Test:  {len(test_df)} rows")
print(f"Val:   {len(val_df)} rows")

print("\nüìä Summary by SEQUENCE:")
print(summary_df.to_string(index=False))
```
